```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import numpy as np
import seaborn as sns
```

```python
dataset = [['Milk', 'Onion', 'Apple', 'Kidney Beans', 'Eggs', 'Yogurt'],
['Banana', 'Onion', 'Apple', 'Kidney Beans', 'Eggs', 'Yogurt'],
['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
['Milk', 'Ice cream', 'Corn', 'Kidney Beans', 'Yogurt'],
['Corn', 'Apple', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
dataset
```

```python
#Transaction Encoder
# chuyển mặt hàng thành từng cột
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df
```

```python
# Tìm tập mục phổ biến
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
#Mining rules from frequent itemset
#Tìm luật kết hợp và lọc theo độ tin cậy
rules=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.85)
```
