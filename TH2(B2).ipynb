```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
# đọc file ,hiển thị dữ liệu
df = pd.read_csv('ThiTHPT2018.csv')
df
```

```python
#xem có bn cột
df.columns
```

```python
#xem thông tin chung
df.info()
```

```python
#Tính trung bình của các môn Toán, Văn. Anh sx theo thứ tự, vẽ đồ thị bar
df[['Toan','Van','Anh','Ly','Hoa','Sinh','Su','Dia','GDCD']].mean().sort_values(ascending= False).plot(kind='bar')
plt.gcf().set_size_inches(18,9)
```

```python
# tính trung vị các môn
df[['Toan','Van','Anh','Ly','Hoa','Sinh','Su','Dia','GDCD']].median()
```

```python
#Tính độ lệch chuẩn
df[['Toan','Van','Anh','Ly','Hoa','Sinh','Su','Dia','GDCD']].std()
```

```python
# vẽ bđ hist
df[['Toan','Van','Anh','Ly','Hoa','Sinh','Su','Dia','GDCD']].hist()
plt.gcf().set_size_inches(20,8)
```

```python
df['Toan'].hist(bins=300)
plt.gcf().set_size_inches(20,8)
#để trang trí có thể sửa lại
plt.xlabel('Toan')
plt.ylabel('Frequency')
plt.title("Toan Histogram")
```

```python
# cho biết đọ lệch gần vs 0 thì ko lệch, âm là lệch trái ,dương lệch phải 
df[['Toan','Van','Anh','Ly','Hoa','Sinh','Su','Dia','GDCD']].skew()
```

```python
# vẽ đồ thị boxplot
df[['Toan','Van','Anh','Ly','Hoa','Sinh','Su','Dia','GDCD']].plot(kind='box')
```

```python
# vẽ bđ hệ số tương quan
cor = df[['Toan','Van','Anh','Ly','Hoa','Sinh','Su','Dia','GDCD']].corr()
# hiển thị chỉ số tương quan
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.gcf().set_size_inches(15,8)
```

```python
df_HN_HCM_DN[['Toan','Van','Anh']].plot(kind='box')
```
