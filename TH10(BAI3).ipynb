```python
import pandas as pd # working with data
import numpy as np # working with arrays
import matplotlib.pyplot as plt # visualization
import seaborn as sb # visualization
from mpl_toolkits.mplot3d import Axes3D # 3d plot
from termcolor import colored as cl # text customization

from sklearn.preprocessing import StandardScaler # data normalization
from sklearn.cluster import KMeans # K-means algorithm

plt.rcParams['figure.figsize'] = (20, 10)
sb.set_style('whitegrid')

# IMPORTING DATA
```

```python
df = pd.read_csv('Mall_Customers.csv')
#df.drop('Unnamed: 0', axis = 1, inplace = True)
df.set_index('CustomerID', inplace = True)
```

```python
df
```

```python
# PhÃ¢n tÃ­ch dá»¯ liá»‡u

# Xem sá»± phÃ¢n bá»‘ cá»§a trÆ°á»ng tuá»•i

print(cl(df['Age'].describe(), attrs = ['bold']))

sb.distplot(df['Age'], 
            color = 'orange')
plt.title('AGE DISTRIBUTION', 
          fontsize  = 18)
plt.xlabel('Age', 
           fontsize = 16)
plt.ylabel('Frequency', 
           fontsize = 16)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.savefig('age_distribution.png')
plt.show()
```

```python
# PhÃ¢n tÃ­ch dá»¯ liá»‡u

# Xem sá»± phÃ¢n bá»‘ cá»§a trÆ°á»ng Annual Income

print(cl(df['Annual Income (k$)'].describe(), attrs = ['bold']))

sb.distplot(df['Annual Income (k$)'], 
            color = 'orange')
plt.title('Annual Income (k$) DISTRIBUTION', 
          fontsize  = 18)
plt.xlabel('Annual Income (k$)', 
           fontsize = 16)
plt.ylabel('Frequency', 
           fontsize = 16)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.savefig('AnnualIncome(k$)_distribution.png')
plt.show()
```

```python
# PhÃ¢n tÃ­ch dá»¯ liá»‡u

# Xem sá»± phÃ¢n bá»‘ cá»§a trÆ°á»ng Spending Score 

print(cl(df['Spending Score (1-100)'].describe(), attrs = ['bold']))

sb.distplot(df['Spending Score (1-100)'], 
            color = 'orange')
plt.title('Spending Score (1-100) DISTRIBUTION', 
          fontsize  = 18)
plt.xlabel('Spending Score (1-100)', 
           fontsize = 16)
plt.ylabel('Frequency', 
           fontsize = 16)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.savefig('SpendingScore(1-100)_distribution.png')
plt.show()
```

```python
# cÃ³ thá»ƒ xem cÃ³ bn nam (1) vÃ  bn ná»¯ (0)

sb.countplot(df['Genre'],
            palette = ['coral','deepskyblue'],
            edgecolor = 'darkgrey')
plt.title('Credit card Male cases(1) and Female cases(0)', 
          fontsize = 18)
plt.xlabel('Male value',
          fontsize = 18)
plt.ylabel('Number of people',
          fontsize = 16 )
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.savefig('Genre_cases.png')
plt.show()
```

```python
#phÃ¢n tÃ­ch sá»± tÆ°Æ¡ng quan giá»¯a Age vÃ  Annual Income (k$)
sb.scatterplot('Age','Annual Income (k$)', data=df,
              color = 'deepskyblue',
              s = 150, alpha = 0.6,
              edgecolor = 'b')
plt.title('AGE/ANNUAL INCOME (K$) ', fontsize = 18)
plt.xlabel('Age', fontsize = 16)
plt.ylabel('Annual Income (k$)',fontsize = 16)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.savefig('age_annualincome.png')
plt.show()
```
