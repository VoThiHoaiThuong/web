```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = pd.read_csv('EmployeeSalary.csv')
data
```

```python
# kt xem cụ thể những dòng thiếu là những dòng nào
df=data.drop(['ID'], axis=1)
```

```python
# chuẩn hóa minmax các trường dữ liệu 
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms.fit(df)
data_mms= mms.transform(df)
```

```python
#đưa vào dataframe
data_mms=pd.DataFrame(data_mms, columns=['WorkingYears','Salary'])
data_mms.headad(10)
```

```python
# Rời rạc hóa dl
cut_labels= ['Low','Medium','High']
cut_bins=[0,800,1200,5000] #0:min, 5000:max
df['Sal_Cat'] = pd.cut(df['Salary'], bins=cut_bins, labels=cut_labels)
df.head(10)
```
