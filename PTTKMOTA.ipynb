```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('ThiTHPT2018.csv')
data
```

```python
# xem bn dong va cot
data.shape
```

```python
#xem tên cột
data.columns
```

```python
data.info()
```

```python
data[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh',
       'Su', 'Dia', 'GDCD']].mean().sort_values(ascending = False).plot(kind='bar')
plt.gcf().set_size_inches(20,8)
```

```python
# tính trung vị 
data[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh',
       'Su', 'Dia', 'GDCD']].median()
```

```python
#tính tần số xuất hiện
data[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh',
       'Su', 'Dia', 'GDCD']].mode()
```

```python
#xem điểm tb theo từng tỉnh thành
data.groupby('Ten Tinh')[['Toan', 'Van', 'Anh']].mean().plot(kind ='bar')
plt.gcf().set_size_inches(20,8)
```

```python
# điểm trung bình của môn toán của cả nước
data.groupby('Ten Tinh')['Toan'].mean().sort_values(ascending = False)
```

```python
# độ lệch chuẩn
data[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh',
       'Su', 'Dia', 'GDCD']].std()
```

```python
data[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh',
       'Su', 'Dia', 'GDCD']].hist()
plt.gcf().set_size_inches(20,8)
```

```python
data['Toan'].hist(bins=300)
plt.gcf().set_size_inches(20,8)
plt.xlabel('Toan')
plt.ylabel('Frequency')
plt.title('Toan Histogram')
```

```python
# cho ra độ lệch
data[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh',
       'Su', 'Dia', 'GDCD']].skew()
```

```python
# tính độ nhọn
data[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh',
       'Su', 'Dia', 'GDCD']].kurt()
```

```python
data[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh',
       'Su', 'Dia', 'GDCD']].plot(kind ='box')
plt.gcf().set_size_inches(20,8)
```

```python
# tính tứ phân vị
data[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh',
       'Su', 'Dia', 'GDCD']].quantile(0.5)
```

```python
# hệ số tương quan 
data[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh',
       'Su', 'Dia', 'GDCD']].corr()
```

```python
cor = data[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh',
       'Su', 'Dia', 'GDCD']].corr()
sns.heatmap(cor , annot= True, cmap=plt.cm.Reds)
plt.gcf().set_size_inches(20,8)
```

```python
# bđ scatter mô tả giữa 2 biến định lượng
data.plot(kind='scatter', x='Ly',y='Toan')
```
