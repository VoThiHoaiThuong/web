```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import numpy as np
import seaborn as sns
data= pd.read_csv('Market_Basket_AR.csv', usecols = range(10))
data.head(10)
```

```python
#stack() được sử dụng để chuyển đổi các giá trị trong dataframe thành dạng Series,apply(pd.Series) được sử dụng để áp dụng stack() cho từng cột trong dataframe
#sum(level=0) được sử dụng để tính tổng số lần xuất hiện của mỗi mặt hàng trong các giao dịch.
#get_dummies() được sử dụng để chuyển đổi dữ liệu về dạng 0,1.
df = data.copy()
df = pd.get_dummies(data.apply(pd.Series).stack()).sum(level=0)

df
```

```python
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df1 = pd.DataFrame(te_ary, columns=te.columns_)
df1
```

```python
# tìm tập mục phổ biến 
frequent_itemsets = apriori(df1, min_support = 0.3,use_colnames = True )
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.4)
```
