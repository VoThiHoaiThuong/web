```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df= pd.read_csv('Dlht.csv')
df
```

```python
# Chuyển cột Gender thành dạng nhị phân
df['Gender_Cat'] = df['Gender'].map({'Male':1, 'Famale':0})
```

```python
#Chuyển các giá tri của trường skill về dạng 0, 1
df_dummies = pd.get_dummies(data= df, columns=['Skill'])
df_dummies.head()
```

```python
#Rời rạc hóa trường Salary thành 3 mức: 'Low','Medium','High'
cut_labels = ['Low','Medium','High']
cut_bins = [0,800, 1200, 2000] #0:min , 2000:max
df['Sal_Cat'] = pd.cut(df['Salary'], bins= cut_bins, labels= cut_labels)
df
```

```python
#Chuyển trường OverseaProject về dạng 0, 1 với {'Yes':1, 'No':0}
df['OverseaProject_Cat'] = df['OverseaProject'].map({'Yes':1, 'No':0})
df
```

```python
#Xóa trường Gender và OverseaProject
del df['Gender']
df
```

```python
del df['OverseaProject']
df
```

```python
#Rời rạc hóa trường TrainedYear thành các khái niệm ‘College', 'University',
'Master_PhD'
cut_Train_labels = ['College','University','Master_phD']
cut_Train_bins = [0,4,5, 10] #0:min , 2000:max
df['Train_Cat'] = pd.cut(df['TrainedYear'], bins= cut_Train_bins, labels= cut_Train_labels)
df
```
