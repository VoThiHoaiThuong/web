```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import zscore
import math

df = pd.read_csv('thiTHPT2018.csv')
```

```python
df.drop('Unnamed: 1',axis=1 ,inplace=True)
df
```

```python
df.isnull().sum()
```

```python
file_to_process = ['Toan','Van', 'AnhVan','TenTinh']
```

```python
#Thay thế giá trị thiếu của trường Toán bằng giá trị trung bình 
data = df.copy()
data.Toan.fillna(data.Toan.mean(), inplace=True)
```

```python
#Thay thế giá trị thiếu của trường Toán bằng giá trị thường xuyên
data.Van.fillna(data.Van.mode()[0], inplace=True)
```

```python
#Anh văn thay thế bằng giá trị thường xuyên theo trường tên tỉnh.
anh_mode = data.groupby('Ma Tinh')['Anh'].apply(lambda x:x.mode()[0]).to_dict()
data['Anh'] = data.apply(lambda x: anh_mode[x['Ma Tinh']] if pd.isnull(x['Anh']) else x['Anh'],axis =1)
```

```python
data.isnull().sum()
```

```python
data
```

```python
#Các trường còn lại thay thế bằng giá trị “Khong thi”
data['Ly'] .fillna('Khong thi',inplace=True)
data['Hoa'] .fillna('Khong thi',inplace=True)
data['Sinh'] .fillna('Khong thi',inplace=True)
data['Su'] .fillna('Khong thi',inplace=True)
data['Dia'] .fillna('Khong thi',inplace=True)
data['GDCD'] .fillna('Khong thi',inplace=True)
data['KhoiA'] .fillna('Khong thi',inplace=True)
data['KhoiB'] .fillna('Khong thi',inplace=True)
data['KhoiC'] .fillna('Khong thi',inplace=True)
data['KhoiD'] .fillna('Khong thi',inplace=True)
data['KhoiA1'] .fillna('Khong thi',inplace=True)
data.isnull().sum()
```

```python
#Tính toán tổng số giá trị thiếu của từng cột và tổng số dòng trong tệp dữ liệu:
data1= df.copy()
num_rows = len(data1)
null_values = data1.isnull().sum()
```

```python
#Ví dụ xóa các cột dữ liệu có số lỗi >=50%
#Xóa những cột có số gá trị thiếu >=50%
columns_to_drop = []
for col in null_values.index:
    if null_values[col] / num_rows >= 0.5:
        columns_to_drop.append(col)
data1
```

```python
# Điền giá trị 0 cho những ô bị thiếu
data1.fillna(0, inplace=True)
data1
```

```python
data1.replace(0, pd.np.nan, inplace=True)
data1
```

```python
for col in data1.columns:
    if col not in columns_to_drop:
        null_percentage = null_values[col] / num_rows
        if null_percentage <0.2:
            if data1[col].dtype == 'object':
                common_value = data1[col].mode()[0]
                data1[col].fillna(common_value, inplace=True)
            else:
                mean_value = data1[col].mean()
                data1[col].fillna(mean_value, inplace=True)
data1
```

```python
# xóa những dòng có giá trị thiếu 
data3= data.copy()
data3.dropna(inplace=True)
data3.isnull().sum()
```

```python
# chuẩn hóa theo zcore tính khoảng lệch cao nhất và thấp nhất so với tb bn độ lệch chuẩn
df_ly=data3['Ly']
z_score_ly=zscore(df_ly) #Tính z-score
z_score_ly #xuất ra màn hình
```

```python
# chuẩn hóa theo zcore tính khoảng lệch cao nhất và thấp nhất so với tb bn độ lệch chuẩn
df_toan=data3['Toan']
z_score_toan=zscore(df_toan) #Tính z-score
z_score_toan #xuất ra màn hình
```
