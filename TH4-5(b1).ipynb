```python
df.isnull().sum()
```

```python
df1=df[['WorkingYear','Salary']]
```

```python
# chuẩn hóa dữ liệu trong khoảng không bầu(Normalization): X_mms=(X-x_min)/(x_max-x_min)
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms.fit(df1)
df_mms=mms.transform(df1)
```

```python
#Convert to Dataframe
df_mms = pd.DataFrame(df_mms, columns=['WorkingYears', 'Salary'])
df_mms.head(10)
```

```python
# Vẽ đồ thị dưới dạng max_min về dạng 0 1
df_mms.plot(kind='box')
plt.title('data after max_min normalization')
```

```python
# Tính khoảng lệch cao nhất và thấp nhất của biến so với TB (z score)
#Standardization : x_stand =(x-mean(x))/StandarDeviation
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df_stand = sc.fit_transform(df1)
```

```python
#Convert to Dataframe
df_stand = pd.DataFrame(df_stand, columns=['WorkingYears','Salary'])
df_stand.describe()
```

```python
#Lệch bn so vs TB độ lệch chuẩn
df_stand.head(10)
```

```python
#Vẽ boxplot
df_stand.plot(kind='box')
plt.title('data after standardization')
```
