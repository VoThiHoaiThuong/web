```python
# xd mô hình dự báo lương biến phụ thuộc
# Salary(Y) = b +a1*Gender +a2*Age + a3*WorkingYear+ a4*TrainedYear
X =df[['Gender','Age','WorkingYear','TrainedYear']] #biến độc lập
Y =df['Salary'] #biến phụ thuộc
```

```python
X #biến định danh phân loại
```

```python
# xử lý biến giả (dummy) Gender(Male:1,Famale:0)
X=pd.get_dummies(X,columns=['Gender'],drop_first=True)
```

```python
X
```

```python
# xd mô hình hồi quy
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(X,Y)
```

```python
# tham số mô hình hồi quy tuyến tính bội
print('R square:', lm.score(X,Y)) # R^2
print('Slope:',lm.coef_) # các hệ số hồi quy 
print('intercept:',lm.intercept_ ) #hệ số chẵn
```

```python
print('Number of Coefficients:', len(lm.coef_))
```

```python
pd.DataFrame(lm.coef_,X.columns)
```

```python
# Salary = 494.9904 +39.0577*Gender +33.6526*WorkingYear+153.50*TrainedYear	-4.30399*Age
# giả sử có 1 người nam 30 tuổi, số năm đào tạo 4 năm ,làm việc ở cty là 4 năm
494.9904 +39.0577*1 +33.6526*3+153.50*4-4.30399*30
```

```python
# dự báo lương
Y_pred=lm.predict(X)
Y_pred=pd.DataFrame(Y_pred, columns=['Salary_predict'])
# so sánh với lương thực tế
result_compare = pd.concat([X,Y,Y_pred], axis=1)
result_compare['Deviation'] = result_compare['Salary_predict']-result_compare['Salary']
result_compare
```

```python
# tổng lương kq dự báo rất gần với ko
# dự báo lương
result_compare['Deviation'].sum()
```
